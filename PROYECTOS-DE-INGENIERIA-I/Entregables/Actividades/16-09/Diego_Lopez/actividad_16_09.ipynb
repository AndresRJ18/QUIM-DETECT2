```python
# Carga solo df1 y df2 para entrenar
df = pd.concat([df1, df2])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df = df.sort_values('Date').reset_index(drop=True)
df.head()
```

```python
df
```

```python
import pandas as pd

# Extraer año y mes
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Crear columnas one-hot
df = pd.get_dummies(df, columns=['Year', 'Month'], prefix=['Year', 'Month'])

print(df.head())
```

```python
df['day'] = df['Date'].dt.day
```

```python
import numpy as np

df['day_sin'] = np.sin(2 * np.pi * df['day'] / 31)
df['day_cos'] = np.cos(2 * np.pi * df['day'] / 31)
```

```python
df.columns
```

```python
X = df[
    [
        'Site ID',
        'Daily AQI Value',
        'Method Code',
        'CBSA Code',
        'State FIPS Code',
        'County FIPS Code',
        'Site Latitude',
        'Site Longitude',
        'Year_2023', 'Year_2024',
        'Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6',
        'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11', 'Month_12',
        'day_sin', 'day_cos'   # 👈 añadidas
    ]
]

# Variable objetivo
y = df['Daily Mean PM10 Concentration']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenar X e y para calcular correlación
data_corr = X.copy()
data_corr['target'] = y

# Calcular matriz de correlación
corr_matrix = data_corr.corr()

# Dibujar el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Mapa de calor de correlaciones")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Lista de columnas numéricas/booleanas útiles
cols_num_bool = [
    'Daily AQI Value',
    'Site Latitude', 'Site Longitude',
    'Year_2023', 'Year_2024',
    'Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6',
    'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11', 'Month_12',
    'day_sin', 'day_cos'   # 👈 añadimos día cíclico
]

# Crear DataFrame para la correlación
corr_df = X[cols_num_bool].copy()
corr_df['Daily Mean PM10 Concentration'] = y

# Calcular correlación
corr = corr_df.corr()

# Graficar mapa de calor
plt.figure(figsize=(14,10))
sns.heatmap(corr, annot=True, fmt=".2f", linewidths=0.5, cmap='coolwarm', cbar=True)
plt.title('Mapa de calor de correlación (variables numéricas, booleanas y día cíclico)')
plt.show()
```

```python
# Calcular matriz de correlación para corr_df
corr_matrix = corr_df.corr()

# Extraer la columna del target y tomar el valor absoluto para comparar magnitudes
target_corr = corr_matrix['Daily Mean PM10 Concentration'].abs()

# Ordenar de mayor a menor y excluir la correlación perfecta consigo misma
target_corr = target_corr.drop('Daily Mean PM10 Concentration').sort_values(ascending=False)

# Mostrar resultados
print(target_corr)

# Si quieres solo las variables con correlación > cierto umbral, por ejemplo 0.3
umbral = 0.3
variables_fuerte_corr = target_corr[target_corr > umbral].index.to_list()

print("Variables con alta correlación (>|0.3|):", variables_fuerte_corr)
```

```python
# Para X (variables predictoras)
print("Tipos de datos en X:")
print(X.dtypes)
```

```python
# Para y (variable objetivo)
print("\nTipo de dato en y:")
print(y.dtype)
```

```python
# Listado de variables categóricas para One-Hot Encoding
categorical_vars = ['Site ID', 'Method Code', 'CBSA Code', 'State FIPS Code', 'County FIPS Code']

# One-Hot Encoding con pandas
X_encoded = pd.get_dummies(X, columns=categorical_vars, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Solo variables numéricas continuas
num_vars = ['Daily AQI Value', 'Site Latitude', 'Site Longitude', 'day_sin', 'day_cos']

scaler = StandardScaler()
X_encoded[num_vars] = scaler.fit_transform(X_encoded[num_vars])

print("Escalado aplicado a:", num_vars)

```

```python
print(X_encoded.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.3, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"R^2: {r2}")
```

```python
# Paso 1: Fecha
df3['Date'] = pd.to_datetime(df3['Date'], format='%m/%d/%Y')

# Paso 2: Dummy años (como dijiste, para 2025 ambas 0)
df3['Year_2023'] = 0
df3['Year_2024'] = 0

# Paso 3: Dummy meses
df3['Month'] = df3['Date'].dt.month
for m in range(1, 13):
    df3[f'Month_{m}'] = (df3['Month'] == m).astype(int)
df3.drop('Month', axis=1, inplace=True)

# Paso 4: Selección de variables (antes de codificar)
vars_para_modelo = [
    'Site ID', 'Daily AQI Value', 'Daily Obs Count', 'Percent Complete', 'Method Code',
    'CBSA Code', 'State FIPS Code', 'County FIPS Code', 'Site Latitude', 'Site Longitude',
    'Year_2023', 'Year_2024'
] + [f'Month_{m}' for m in range(1,13)]

df3_model = df3[vars_para_modelo]

# Paso 5: One-Hot Encoding para las variables categóricas (igual que en entrenamiento)
categorical_vars = ['Site ID', 'Method Code', 'CBSA Code', 'State FIPS Code', 'County FIPS Code']
df3_encoded = pd.get_dummies(df3_model, columns=categorical_vars, drop_first=True)

# Paso 6: Asegurarse que df3_encoded tenga las mismas columnas que X_encoded (modelo entrenado)
for col in X_encoded.columns:
    if col not in df3_encoded.columns:
        df3_encoded[col] = 0  # columnas faltantes agregadas con 0

# Eliminar columnas extras si las hay (por si aparece alguna nueva)
df3_encoded = df3_encoded[X_encoded.columns]

# Paso 7: Escalar variables numéricas usando el scaler ya entrenado
num_vars = ['Daily AQI Value', 'Site Latitude', 'Site Longitude', 'day_sin', 'day_cos']

df3_encoded[num_vars] = scaler.transform(df3_encoded[num_vars])

# Ahora df3_encoded está listo para hacer predicciones con tu modelo
```

```python
# Predecir PM10 para 2025
y_pred_2025 = model.predict(df3_encoded)
```

```python
# Los valores reales de PM10 en 2025
y_real_2025 = df3['Daily Mean PM10 Concentration'].values

# Crear un DataFrame con ambos para comparar
comparacion = pd.DataFrame({
    'Real': y_real_2025,
    'Predicho': y_pred_2025
})

# Mostrar primeras filas para revisar
print(comparacion.head())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_real_2025, y_pred_2025)
mae = mean_absolute_error(y_real_2025, y_pred_2025)
r2 = r2_score(y_real_2025, y_pred_2025)

print(f'MSE: {mse:.3f}')
print(f'MAE: {mae:.3f}')
print(f'R2: {r2:.3f}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(comparacion['Real'].values, label='Real')
plt.plot(comparacion['Predicho'].values, label='Predicho', alpha=0.7)
plt.legend()
plt.title('Comparación: Valores reales vs Predichos de PM10 (2025)')
plt.show()
```
