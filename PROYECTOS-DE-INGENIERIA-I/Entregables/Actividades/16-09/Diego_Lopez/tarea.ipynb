```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# =========================
# 1. Leer datos
# =========================
df1 = pd.read_csv("datos_2023.csv")
df2 = pd.read_csv("datos_2024.csv")
df3 = pd.read_csv("datos_2025.csv")

# Concatenar todo
df_all = pd.concat([df1, df2, df3], ignore_index=True)

# =========================
# 2. Procesar fechas
# =========================
df_all["Date"] = pd.to_datetime(df_all["Date"])
df_all["year"] = df_all["Date"].dt.year
df_all["month"] = df_all["Date"].dt.month
df_all["dayofweek"] = df_all["Date"].dt.dayofweek   # lunes=0
df_all["dayofyear"] = df_all["Date"].dt.dayofyear

# =========================
# 3. OneHotEncoder para Site ID
# =========================
ohe = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")

# Entrenar el encoder SOLO con datos 2023‚Äì2024
ohe.fit(pd.concat([df1, df2])[["Site ID"]])

# Transformar TODO el df_all (incluyendo 2025)
encoded_sites_all = ohe.transform(df_all[["Site ID"]])
encoded_df_all = pd.DataFrame(encoded_sites_all, columns=ohe.get_feature_names_out(["Site ID"]))

# Concatenar al df
df_all = pd.concat([df_all.reset_index(drop=True), encoded_df_all], axis=1)

# =========================
# 4. Crear variables de retraso (lags)
# =========================
df_all = df_all.sort_values(by=["Site ID", "Date"])

for lag in [1, 7, 30]:
    df_all[f"PM10_lag{lag}"] = df_all.groupby("Site ID")["Daily Mean PM10 Concentration"].shift(lag)

# =========================
# 5. Variables c√≠clicas de tiempo
# =========================
df_all["month_sin"] = np.sin(2 * np.pi * df_all["month"] / 12)
df_all["month_cos"] = np.cos(2 * np.pi * df_all["month"] / 12)
df_all["dow_sin"]   = np.sin(2 * np.pi * df_all["dayofweek"] / 7)
df_all["dow_cos"]   = np.cos(2 * np.pi * df_all["dayofweek"] / 7)
df_all["doy_sin"]   = np.sin(2 * np.pi * df_all["dayofyear"] / 365)
df_all["doy_cos"]   = np.cos(2 * np.pi * df_all["dayofyear"] / 365)

# =========================
# 6. Definir features
# =========================
features = [
    "PM10_lag1", "PM10_lag7", "PM10_lag30",
    "month_sin", "month_cos",
    "dow_sin", "dow_cos",
    "doy_sin", "doy_cos"
] + list(ohe.get_feature_names_out(["Site ID"]))

# =========================
# 7. Eliminar filas con NaN en los features (ej: primeros d√≠as de cada sitio)
# =========================
df_all = df_all.dropna(subset=features)

# =========================
# 8. Separar train, test, futuro
# =========================
train = df_all[df_all["year"] == 2023]
test = df_all[df_all["year"] == 2024]
future = df_all[df_all["year"] == 2025]

X_train, y_train = train[features], train["Daily Mean PM10 Concentration"]
X_test, y_test   = test[features], test["Daily Mean PM10 Concentration"]
X_2025           = future[features]

# =========================
# 9. Entrenar modelo
# =========================
model = LinearRegression()
model.fit(X_train, y_train)

# =========================
# 10. Evaluar en 2024
# =========================
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Resultados en 2024:")
print("MSE:", mse)
print("RMSE:", rmse)
print("R¬≤:", r2)

# =========================
# 11. Predecir 2025
# =========================
y_pred_2025 = model.predict(X_2025)
future["PM10_pred"] = y_pred_2025

# Guardar predicciones 2025
future.to_csv("predicciones_2025.csv", index=False)
print("Predicciones 2025 guardadas en predicciones_2025.csv")
# =========================
# 12. Evaluar en 2025 (si hay datos reales)
# =========================
if "Daily Mean PM10 Concentration" in future.columns:
    y_true_2025 = future["Daily Mean PM10 Concentration"]
    mse_2025 = mean_squared_error(y_true_2025, y_pred_2025)
    rmse_2025 = np.sqrt(mse_2025)
    r2_2025 = r2_score(y_true_2025, y_pred_2025)

    print("\nResultados en 2025:")
    print("MSE:", mse_2025)
    print("RMSE:", rmse_2025)
    print("R¬≤:", r2_2025)

```
